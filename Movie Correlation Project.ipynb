```python
# import libraries 

import pandas as pd
import seaborn as sns 
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plot we will create


# Read in the data

df = pd.read_csv(r'C:\Users\edkou\OneDrive\Documents\movies.csv')
```

```python
# looking at the data

df.head()
```

```python
# Looking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# Changing the data type column

df['gross'] = df['gross'].fillna(0).astype('int64')
df['budget'] = df['budget'].fillna(0).astype('int64')
```

```python
df.head()
```

```python
# Creating the correct year column

df['yearcorrect'] = df['released'].astype(str).str.split().str[2]

df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False).head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Dropping any duplicate data

df['company'].drop_duplicates().sort_values(ascending=False).head()
```

```python
# Making a scatter plot between the budget vs gross 

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for film')

plt.ylabel('Gross Earnings') 

plt.show()
```

```python
# Looking at the company correlation

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features') 

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
